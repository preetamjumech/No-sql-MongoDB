## 2.3 5 different queries specifying the selection conditions and projection conditions

```
table = db.une
c1 = table.find({},{"Country":1,"Year" : 1,"_id":0,'Area (square kilometres)':1,'Total population':1,
                    "Persons killed in road accidents":1,"Total length of railway lines (km)":1,
                    "Import of goods and services, per cent of GDP":1,
                    'Population aged 0-14, male':1,'Population aged 0-14, female':1}).limit(1000)
pr=[]
for crnt in c1:
    print(crnt)
    country=crnt["Country"]
    year=crnt["Year"]
    area=crnt['Area (square kilometres)']
    population=crnt['Total population']
    person_killed=crnt["Persons killed in road accidents"]
    length_railway=crnt["Total length of railway lines (km)"]
    goods_services=crnt["Import of goods and services, per cent of GDP"]
    male_pop=crnt['Population aged 0-14, male']
    female_pop=crnt["Population aged 0-14, female"]
    info=(country,year,area,population,person_killed,length_railway,goods_services,male_pop,female_pop)
    pr.append(info)
```

```
c1 = db.une.find({},{"_id":0}).limit(10)
for crnt in c1:
    print(crnt)
```

## Selection

```
c1 = db.une.find({"Persons killed in road accidents":{"$gt" : 200}},{"_id":0}).limit(5)
for crnt in c1:
    print(crnt)
```

```
c1 = db.une.find({"Area (square kilometres)":{"$gt" : 28000}},{"_id":0}).limit(5)
for crnt in c1:
    print(crnt)
```

```
c1 = db.une.find({"Year": '2000'},{"_id":0}).limit(5)
for crnt in c1:
    print(crnt)
```

```
c1 = db.une.find({"Female tertiary students, percent of total": {"$gt" : 55}},{"_id":0}).limit(6)
for crnt in c1:
    print(crnt)
```
